```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
import configparser
```

```python
configParser = configparser.RawConfigParser()   
configFilePath = r'./data/config.txt'
configParser.read(configFilePath)

n_x = int(configParser.get('config-file', 'n_x'))
n_y = int(configParser.get('config-file', 'n_y'))

framesTotal = int(configParser.get('config-file', 'frames'))

title = configParser.get('config-file', 'title')

fileprefix = "./data/" + title

fig = plt.figure()
# fig.suptitle(title, fontsize=20)

b = np.random.random((n_x, n_y))

a = np.random.random((n_x, n_y))
im1 = plt.imshow(a, cmap='plasma', interpolation='bicubic')

alphas = np.loadtxt(fileprefix + "_potential.txt").transpose()
im2 = plt.imshow(b, cmap='hot', alpha=alphas)

plt.axis('off')

plt.close()

def init():
    im1.set_data(np.zeros((n_x, n_y)))
    return [im1, im2]

def animate(i):
    file = fileprefix + "_frame" + "{}".format(i+1).zfill(5) + ".txt"
    data=np.loadtxt(file)
    im1.set_data(data.transpose())
    
    im2.set_data(alphas)
    
    # autoscale mozna zostawic ale mozna rowniez nie
    im1.autoscale()
    return [im1, im2]

anim = animation.FuncAnimation(fig, animate, frames=framesTotal, interval=100, blit=True)
HTML(anim.to_html5_video())
```
